```python
# 막대 그래프
member_IDs = ['m_01','m_02', 'm_03','m_04']
before_ex = [27,35,40,33]
after_ex = [30, 38, 42, 37]
```

```python
n_data = len(member_IDs)
index = np.arange(n_data)
plt.bar(index, before_ex)
plt.xticks(fontname='Arial')
plt.show()
```

```python
# 그래프 요소 틱 레이블 넣기 
plt.bar(index, before_ex, tick_label=member_IDs)
plt.xticks(fontname='Arial')
plt.show()
```

```python
# 각각 멤버별로 컬러 지정
colors = ['r','g','b','m']
plt.bar(index, before_ex, tick_label=member_IDs, color=colors)
plt.xticks(fontname='Arial')
plt.show()
```

```python
# 막대 그래프 두꼐 조절
colors = ['r','g','b','m']
plt.bar(index, before_ex, tick_label=member_IDs, color=colors, width=0.6)
plt.xticks(fontname='Arial')
plt.show()
```

```python
colors = ['r','g','b','m']
plt.barh(index,before_ex, tick_label=member_IDs, color=colors)
plt.xticks(fontname='Arial')
plt.show()
```

```python
#운동 전후 비교 그래프
barWidth =0.4
plt.bar(index,before_ex, color='c', align='edge',width=barWidth, label='before')
plt.bar(index + barWidth, after_ex, color='m', align='edge',width=barWidth,label='after')
plt.xticks(index + barWidth, member_IDs)
plt.legend()
plt.xlabel('회원ID')
plt.ylabel('윗몸 일으키기')
plt.title('운동시작 전/후 근지구력 변화 비교')
plt.show()
```

```python
# 파이는 3시 방향이 Start 지점이며, 반시계 방향으로 돈다
fruit = ['사과','바나나','딸기','오렌지','포도']
result = [7,6,3,2,2]
plt.pie(result)
plt.show()
```

```python
plt.pie(result, labels=fruit, autopct='%.1f%%')
# 끝에 %기호를 2개 붙여야 한개 들어간다
plt.show()
```

```python
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90,\
       counterclock = False) # 반시계 방향으로 회전하는 것을 시간방향으로 변경
plt.show()
```

```python
# 특정 부분 강조
explode_value = [0.1,0,0,0,0]
plt.pie(result, labels=fruit, autopct='%.1f%%', startangle=90,counterclock=False, explode=explode_value, shadow=True)
plt.show()
```
