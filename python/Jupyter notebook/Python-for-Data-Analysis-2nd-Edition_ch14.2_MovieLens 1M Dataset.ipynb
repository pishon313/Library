```python
import pandas as pd

pd.options.display.max_rows = 10

unames = ['user_id', 'gender', 'age', 'occupation', 'zip']
users = pd.read_table ('pydata-book-2nd-Edition/datasets/movielens/users.dat', sep="::", header=None, names=unames, encoding = 'cp949')

rnames =['user_id', 'movie_id', 'rating', 'timestamp']
ratings = pd.read_table('pydata-book-2nd-Edition/datasets/movielens/ratings.dat', sep="::", header=None, names=rnames, encoding = 'cp949')


mnames = ['movie_id', 'title', 'genres']
movies = pd.read_table('pydata-book-2nd-Edition/datasets/movielens/movies.dat', sep="::", header=None, names=mnames, encoding = 'cp949')
```
